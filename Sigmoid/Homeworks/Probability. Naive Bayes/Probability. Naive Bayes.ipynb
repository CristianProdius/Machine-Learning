# Probability. Naive Bayes

```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNBdf
```

```python
df =pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}

for column in df.columns:
    label_encoder = LabelEncoder()
    df[column] = label_encoder.fit_transform(df[column])
    label_encoders[column] = label_encoder

df.head()
```

```python
X = df.drop(['Churn'], axis = 1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.3)
```

## Gausian Naive Bayes

```python
Gauss = GaussianNB()
Gauss.fit(X_train, y_train)
Gauss.score(X_test, y_test)
```

## Multinomial Naive Bayes

```python
Multinomial = MultinomialNB()
Multinomial.fit(X_train, y_train)
Multinomial.score(X_test, y_test)
```

## Bernouli Naive Bayes

```python
Bernoulli = BernoulliNB()
Bernoulli.fit(X_train, y_train)
Bernoulli.score(X_test, y_test)
```

```python
scores = [Gauss.score(X_test, y_test), Multinomial.score(X_test, y_test), Bernoulli.score(X_test, y_test)]
table = pd.DataFrame({'ML Model': ['Gauss', 'Multi', 'Bernoulli'], 'Accuracy': scores})

table
```
